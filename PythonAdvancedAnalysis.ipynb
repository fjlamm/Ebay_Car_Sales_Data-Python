now that we have cleaned the data a little bit, I will dig some more into some aspects like, what brands are more expensive for instance

let's look into that by aggregating by brand and calculating both, mean and max per brand, then sorting descending and looking at the top five most expensive brands

```python
autos['brand'].value_counts(normalize='True').sort_values()
```

```python
my_list=[]
max_dict={}
mean_dict={}
my_list=autos['brand'].value_counts().index
```

```python
for var in my_list:
    temp_max=autos['price'][autos['brand']==var].max()
    max_dict[var]=temp_max
    
    temp_mean=autos['price'][autos['brand']==var].mean()
    mean_dict[var]=temp_mean
    
```

```python
sorted(max_dict.items())
 
```

We can see that porsche has the max price. let's see the mean price per brand

```python
sorted(mean_dict.items())
```

Porsche also has the highest mean price. 

Now, as we can see. Using dictionaries it is more difficult to get the top five or even order by value. For both of these processes we would need loops.

Let's see if we can do it easier and faster through Pandas dataframes

```python
temp_df=autos.groupby('brand')['price'].max()
```

```python
temp_df.sort_values(ascending=False).head()
```

as we can see, in these two short steps we:
    got the max price for each brand
    printed the top 5 
    sorted in descending order as we wanted originally.
 
 ### This is the power of PANDAS!!!!!

```python
temp_df=autos.groupby('brand')['price'].mean()
temp_df.sort_values(ascending=False).head()
```
